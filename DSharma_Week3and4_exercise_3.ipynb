```python
import numpy as np
import pandas as pd
```

Create a series and practice basic arithmetic steps

```python
Value1 = 7.3, -2.5, 3.4, 1.5
index1 = 'a', 'c', 'd', 'e'

Series_1 = pd.Series(data=Value1, index=index1)
```

```python
Value2 = -2.1, 3.6, -1.5, 4, 3.1
index2 = 'a', 'c', 'e', 'f', 'g'

Series_2 = pd.Series(data=Value2, index=index2)
```

Add Series 1 and Series 2 together and print the results

```python
Series_3 = Series_1+Series_2
```

```python
print("The total of series 1 and series 2 is below:\n", Series_3)
```

```python
# Using fill_value to get remove NaN values

print("Total of series after using fill_value is below: ")
Series_1.add(Series_2, fill_value=0)
```

Subtract Series 1 from Series 2 and print the results

```python
Series_4 = Series_2 - Series_1
```

```python
print("Final series when we subtract 1 from series 2 is below:\n", Series_4)
```

```python
# Using fill_value to get remove NaN values

print("Final of series after using fill_value is below: ")
Series_2.sub(Series_1, fill_value=0)
```
